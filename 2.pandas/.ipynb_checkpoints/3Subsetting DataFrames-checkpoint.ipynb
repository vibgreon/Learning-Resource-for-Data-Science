```python
# filtering rows using []
# label based indexing using ".loc"
# integer based indexing using ".iloc"
```

```python
import pandas as pd
import seaborn as sns
```

```python
# in this we're using seaborn to use a particular default dataset

df = sns.load_dataset('tips')
df.head()
```

```python
# show values
# panda series

df['tip'].head()
```

```python
# show value in dataframe
# panda dataframe

df[['tip']].head()
```

```python
df[['tip','sex']].head()
```

```python
# select rows 3 to 4

df[3:5]
```

## .loc

```python
# it is label based indexing

df.head()
```

```python
# df.loc[rows, columns]

df.loc[:,'tip'].head()
```

```python
df.loc[:,['tip','time']].head()
```

```python
# here its NOT like last one being included
# its about LABELS

df.loc[1:4,['tip','time']].head()
```

```python
# it will give you the 3rd index value

df.loc[3]
```

```python
df.loc[[3]]
```

```python
# because the index is integer, we are putting integer values

df.loc[[3,6]]
```

```python
df2 = df.set_index('sex')
df2.head()
```

```python
df2.loc['Female']
```

```python
df2.loc['Female',['tip','day','time']]
```

## .iloc

```python
# it is integer based indexing

df.head()
```

```python
# giving rows 1 to 3, last one is not included

df.iloc[1:4]
```

```python
# we cannot use names sinceits all integer type

df.iloc[1:4,[3,4]]
```

```python
df.iloc[1:4,[1,4]]
```
