```python
df = pd.read_csv('auto_clean.csv')
df.head()
```

## Histogram plot with (kde) density estimation

```python
sns.distplot(df['mpg'])
plt.show()
```

```python
sns.distplot(df['mpg'], kde=False, bins=20)
plt.show()
```

```python
plt.figure(figsize=[10,6])
sns.distplot(df['mpg'], kde=False, bins=20)
plt.show()
```

## Scatter Plot

```python
# it does a scatter plot and a fitted linear regression line

sns.lmplot('weight','mpg',data=df)
plt.show()
```

```python
# aspect determines scale of x axis
# size determines scale of y axis

sns.lmplot('weight','mpg',data=df, fit_reg=False, aspect=2, size=6)
plt.show()
```

```python
sns.lmplot('weight','mpg',data=df, fit_reg=False, aspect=2, size=4)
plt.show()
```

```python
# based on certain column, hue classification

sns.lmplot('weight','mpg',data=df, fit_reg=False, aspect=2, size=4, hue='cylinders')
plt.show()
```

```python
sns.lmplot('weight','mpg',data=df, fit_reg=False, aspect=1, size=5, hue='cylinders',col='origin')
plt.show()
```

## KDE Plot (Kernel Distribution Estimation)

```python
# input is array of numbers

sns.kdeplot(df['weight'],df['mpg'], shade=True)
plt.show()
```

## Joint Plot

```python
# scatterplot in middle and histogram on sides

sns.jointplot('weight','mpg',data=df,alpha=0.4)
plt.show()
```

## Pair Grid

```python
# diagonal - histogram density estimation/ histogram
# upper - scatter plot
#lower - 2D KDE plot

g=sns.PairGrid(df, diag_sharey=False)
g.map_lower(sns.kdeplot)
g.map_upper(plt.scatter)
g.map_diag(sns.kdeplot)
```

## Bar plot

```python
sns.countplot(y='cylinders',data=df)
plt.show()
```

## Box Plot

```python
sns.boxplot(x='cylinders',y='mpg',data=df)
plt.show()
```

```python
sns.boxplot(x='origin',y='mpg',data=df,hue='cylinders')
plt.show()
```
