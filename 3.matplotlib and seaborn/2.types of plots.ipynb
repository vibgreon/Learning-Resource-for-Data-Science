```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
col_names = ['mpg','cylinders','dis','hp','weight',
             'acc','year','origin','name']
```

```python
df = pd.read_csv('auto-mpg.data', sep='\s+', names=col_names)
df
```

## Histogram Plot

```python
# Histogram is good for 1D

plt.hist(df['mpg'])
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.hist(df['mpg'], bins=20)
plt.xlabel('mpg')
plt.show()
```

## Scatter Plot

```python
# scatter is use to see the interaction between 2 variables

# alpha value gives transparency to the plots 

plt.figure(figsize=(10,6))
plt.scatter(x=df['weight'],y=df['mpg'],alpha=0.5)

plt.ylabel('mpg')
plt.xlabel('weight')
plt.show()
```

## Hostogram 2D plot

```python
# instead of scatter , it shows plot into bins
# when there are mpre points and they're overlapping each other,
# hist2d is a better choice


plt.figure(figsize=(10,6))
plt.hist2d(x=df['weight'],y=df['mpg'])

plt.ylabel('mpg')
plt.xlabel('weight')
plt.colorbar()
plt.show()
```

## Hexagonal Graph

```python
# gribsize represent size of bins

plt.figure(figsize=(10,6))
plt.hexbin(x=df['weight'],y=df['mpg'], gridsize=(35,15))

plt.ylabel('mpg')
plt.xlabel('weight')
plt.colorbar()
plt.show()
```

```python
# cmap to change color

plt.figure(figsize=(10,6))
plt.hexbin(x=df['weight'],y=df['mpg'], gridsize=(35,15), cmap='coolwarm')

plt.ylabel('mpg')
plt.xlabel('weight')
plt.colorbar()
plt.show()
```
