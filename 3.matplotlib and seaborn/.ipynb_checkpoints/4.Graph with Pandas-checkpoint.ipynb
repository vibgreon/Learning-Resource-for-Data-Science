```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
df = pd.read_csv('auto_clean.csv')
df.tail()
```

## Histogram

```python
df['weight'].plot(kind='hist',bins=20)
```

```python
df.groupby('cylinders')['mpg'].plot(kind='hist',alpha=0.5, legend=True)
```

## Scatter Plot

```python
df.plot(kind='scatter',x='weight',y='mpg')
```

## Box Plot

```python
df.boxplot(column='mpg',by='cylinders',grid=False)
```

## Pivot Table

```python
# this will groupby 'cylinders' and 
# calculate mean for each columns for each cylinder group

dfp = df.pivot_table(index='cylinders',aggfunc='mean')
dfp
```

```python
# dfp.plot(kind='line') will also do the same

dfp.plot.line(y='acc')
```

```python
for col in dfp.columns:
    dfp.plot.line(y=col)
```

## Bar Plot

```python
# .count() will count 'cylinders' in each column
# ['mpg'] will select column 'mpg'

df.groupby('cylinders').count()['mpg'].plot(kind='bar')
plt.show()
```
