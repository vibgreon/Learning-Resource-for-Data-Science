```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
x = np.linspace(0, np.pi*4, num=200)
```

```python
x
```

```python
y = np.sin(x)
```

```python
y
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,'r', label='sin(x)')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.legend()
plt.show()
```

```python
z = np.cos(x)

plt.plot(x,y,'r', label='sin(x)')
plt.plot(x,z,'b', label='cos(x)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.show()
```

```python
plt.plot(x,y,'r', label='sin(x)')
plt.plot(x,z,'b', label='cos(x)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.xlim([0,6])
plt.legend()
plt.show()
```

```python
# subplot(rows, columns, we are goint to use nth plot)

plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(x,z,'b')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(x,z,'b')

plt.tight_layout()
plt.show()
```

```python
# add_axes([x, y, width, height])
# figure is the canvas

fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y,label='sin(x)')
axes.legend()

axes.set_xlabel('x')
axes.set_ylabel('sin(x)')
axes.set_title('matplotlib graph')

plt.show()
```
